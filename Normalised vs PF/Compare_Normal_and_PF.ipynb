# Comparing the difference in normalsing the integrand, integral, or both

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import json
```

## Normalising both integrand and integral

```python
with open('data/both_normalised.json', 'r') as file:
    all_data = json.load(file)
```

```python
# convert each pre-fix notation to it's token length
data_lens = [[[len(data[0][0]), len(data[0][1])], 
              [len(data[1][0]), len(data[1][1])]
             ] for data in all_data]
```

```python
flattened_data = [item for sublist in data_lens for pair in sublist for item in pair]
reshaped_data = [flattened_data[i:i+4] for i in range(0, len(flattened_data), 4)]
df = pd.DataFrame(reshaped_data, columns=['integrand_norm', 'integral_norm', 'integrand_PF', 'integral_PF'])
```

```python
df.head()
```

```python
df_long = pd.DataFrame({
    'integrand': df['integrand_norm'].tolist() + df['integrand_PF'].tolist(),
    'integral': df['integral_norm'].tolist() + df['integral_PF'].tolist(),
    'Dataset': ['Normalised'] * len(df) + ['Partial Fraction'] * len(df)
})
```

```python
# Create a scatter plot using Seaborn with different colors for each dataset
ax = sns.scatterplot(data=df_long, x='integrand', y='integral', hue='Dataset', palette='deep', s=10)
plt.xlabel('# Integrand Tokens', fontsize=14)
plt.ylabel('# Integral Tokens', fontsize=14)
#  plt.title("Normalising Integrand Only", fontsize=16)
# Add a line at y = x
x_vals = np.linspace(0, 600, 100)
plt.plot(x_vals, x_vals, color='red', linestyle='--')
# Set the x-axis and y-axis limits
plt.xlim(0, 600)
plt.ylim(0, 600)
plt.savefig("figures/normalise_both.eps", format='eps')
plt.show()
```
